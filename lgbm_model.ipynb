```python
import gc
gc.collect()
```

```python
import lightgbm as lgb
from tqdm import tqdm
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
pre=pd.DataFrame()
fnl_trn=fnl_trn.reset_index(drop=True)
for e,(i,j) in tqdm(enumerate(kf.split(fnl_trn))):
      train=fnl_trn.loc[i]
      validation=fnl_trn.loc[j]
      mod=lgb.LGBMClassifier(max_bins= 37,
 max_depth= 40,
 min_data_in_leaf= 110,
 min_gain_to_split= 1,
 num_leaves=48)
      mod.fit(train.drop(['isFraud'],1),train['isFraud'])
      pre[e]=list(mod.predict_proba(fnl_tst.drop(['isFraud'],1))[:,1])
      del([train,validation,mod])  
      gc.collect()
```

```python
pre['tot']=pre.mean(axis=1)
ss=pd.read_csv('/kaggle/input/ieee-fraud-detection/sample_submission.csv')
ss['isFraud']=pre['tot']
ss=ss.set_index('TransactionID')
ss.to_csv('sub.csv')
```
